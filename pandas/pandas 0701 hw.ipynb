```python
# seaborn에서 titanic 데이터를 로드하여 데이터프레임(titanic)으로 생성
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic

# # titanic 칼럼별 개수 확인
titanic.value_counts()

# # titanic 전체 기본 정보 확인
titanic.info()

# # titanic 통계 정보 확인
titanic.describe(include='all')
```

```python
# 컬럼별 개수가 300 미만인 칼럼과 데이터타입이
#숫자가 아닌 칼럼은 제외
# 나머지 컬럼을 추출하여 데이터프레임(df) 생성
titanic.value_counts()
cnt = titanic.count() # 칼럼별 데이터 개수 -> 시리즈로 전환
col_list = cnt[ cnt > 300].index # 데이터 개수가 300 이상인 칼럼명만 추출
df = titanic.loc[:,col_list]
df
dtype_list = df.dtypes
(dtype_list == 'int64') | (dtype_list == 'float64')
df = df.loc[:,(dtype_list == 'int64')|(dtype_list == 'float64')]
df
# df 기본 정보
# df.info()
# df의 처음 5개의 정보만
df.head()
# df의 age와 fare 칼럼에서 
#200~ 400번째 자료까지만 추출 df_1
df_1 = df.loc[200:400,['age','fare']]
df_1
#df_1의 age 칼럼을 index 칼럼으로
df_1.set_index('age')
```
