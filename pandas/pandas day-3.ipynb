# 3. pandas에 내장된 그래프 도구 활용


```python
import pandas as pd

df = pd.read_excel('./dataset/남북한발전전력량.xlsx')

df
```

```python
# 0행과 5행의 3열부터 ~ 데이터 가져오기
df_ns  = df.iloc[ [0, 5] , 3: ].copy()
df_ns
                 
```

```python
# 선 그래프 그리기
df_ns.plot()
```

```python
df_ns_t = df_ns.T
df_ns_t.plot()
df_ns_t
```

```python
# column 명 변경 : south와 north 로
df_ns_t.columns = ['south', 'north']
df_ns_t.plot()
```

```python
#  막대 그래프 : df.plot(kind='bar')

df = pd.read_excel('./dataset/남북한발전전력량.xlsx')
df_ns  = df.iloc[ [0, 5] , 3: ]
df_ns.index = ['south', 'north']
# 컬럼명을  int 로 변경 : 1990, 1991
df_ns.columns = df_ns.columns.map(int)

df_ns.info() # datatype이 object 이므로 int형으로 변경
df_ns = df_ns.astype(int)

df_ns_t = df_ns.T
df_ns_t.plot(kind='hist') #  히스토그램 -> df.plot(kind='hist')
df_ns_t.plot(kind='bar')
```
