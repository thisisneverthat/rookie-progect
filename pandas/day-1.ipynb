```python
exam_data = {'수학' : [ 90, 80, 70], '영어' : [ 98, 89, 95],
             '음악' : [ 85, 95, 100], '체육' : [ 100, 90, 90]}
df = pd.DataFrame(exam_data, index=['서준','우현','인아'])
df
```

```python
# 행을 선택 : object.loc[인덱스명] <<인덱스 명으로 선택, 
#             objec.iloc[정수인덱스]<<정수 인덱스로 선택

print(df.loc['서준'])
print()
print(df.iloc[0])

#범위로선택 : object.loc[start : end] <<인덱스 명으로 선택(end 포함)
#             object.iloc[start : end] <<정수 인덱스로 선택(end 미포함)
print(df.loc['서준':'우현']) # 서준과 우현 자료출력
print()
print(df.iloc[0:2]) # 0행과 1행 출력 // 2행 미포함

#서준과 인아의 정보 출력
print(df.loc[['서준','인아']])
print()
print(df.iloc[[0,2]])

#행을 슬라이싱 : object.iloc[시작 인덱스: 끝 인덱스: 간격]
print(df.iloc[0:3:2])
print(df.iloc[: :2])
print(df.iloc[: :-1]) #음수는 뒤에서 부터 출력
```

```python
# object[칼럼명] or objcet.칼럼명
print(df.수학)
print()
print(df[ ['수학','체육'] ]) # 여러 요소를 선택 할경우 [](리스트)안에 []지정
```

```python
# 시리즈 vs 시리즈 연산 : 시리즈 연산자(+,-,*,/) 시리즈
student2 = pd.Series({'국어':90,'영어':80,'수학':100})
student1 + student2
```

```python

add = student1 + student2
sub = student1 - student2
mul = student1 * student2
div = student1 / student2

df =pd.DataFrame([add,sub,mul,div],
                index = ['+','-','*','/'])
df
```
