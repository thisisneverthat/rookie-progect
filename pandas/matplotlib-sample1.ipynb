```python
# %load fonts.py
#!/usr/bin/env python

# In[ ]:

import pandas as pd
import matplotlib.pyplot as plt
# matplotlib 한글 폰트 문제 해결
from matplotlib import rc
import platform 

if platform.system() == 'Darwin':   # Mac 환경 폰트 지정
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    rc('font', family='Malgun Gothic')
    
plt.rc('axes', unicode_minus=False)

```

```python
# '시도별 전출입 인구수.xlsx' 파일을 읽어 df에 저장한 후
# Nan 데이터 처리
# 서울특별시에서 타시로 이동한 인구의 자료만 df_seoul 로 추출
# 인덱스를 전입지 컬럼으로 설정
# 경기도, 충청남도, 강원도, 전라남도의 자료 중 1997년부터 2017년 자료만 추출해서
# df_1 에 저장
# 년도를 인덱스로 바꾸어서 df_t 에 저장
```

```python
# '시도별 전출입 인구수.xlsx' 파일을 읽어 df에 저장한 후
df = pd.read_excel('./dataset/시도별 전출입 인구수.xlsx')
# df.count()
df.iloc[15:25, :]   #  Nan을 이전 값으로 대체
df = df.fillna(method='ffill') # Nan 데이터 처리
df.iloc[15:25, :]

# 서울특별시에서 타시로 이동한 인구의 자료만 df_seoul 로 추출
df_seoul = df[(df['전출지별']=='서울특별시') & (df['전입지별'] !='서울특별시')]
df_seoul

# '전출지별' 의 값이 모두 '서울특별시'로 같으므로 '전출지별' 컬럼 삭제
df_seoul = df_seoul.drop(columns=['전출지별'], axis=1)
df_seoul
```

```python
df_seoul # 전입지별 -> 전입지로 변경후 컬럼을 인덱스로 설정
df_seoul = df_seoul.rename(columns={'전입지별' : '전입지'})
```

```python
df_seoul.set_index('전입지', inplace=True) # 인덱스를 전입지 컬럼으로 설정
```

```python
# 경기도, 충청남도, 강원도, 전라남도의 자료 중 1997년부터 2017년 자료만 추출
col_years = df_seoul.columns
col_years = list(map(str, range(1997, 2018)))  # 1997년부터 2017년 자료만 추출

df_1 = df_seoul.loc[['경상북도', '충청남도', '강원도', '전라남도'], col_years]
df_1

df_t = df_1.T
df_t
```

```python
# 면적 그래프 : 인덱스가 숫자 이어야 함
df_t.index  = df_t.index.map(int)
df_t.index
```

```python
df_t.plot(kind='area', stacked=True, alpha=0.2)
plt.title('서울 -> 타시도 인구 이동', size=12)
plt.ylabel('이동 인구 수')
plt.xlabel('기간')
plt.legend(loc='best')
plt.show()
```
