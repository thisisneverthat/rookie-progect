```python
# 모듈을 프로그램에 삽입. // import 모듈명
import math

print(math.ceil(43.2)) # 올림정수 44
print(math.floor(43.2)) # 내림 정수 43
print(math.sin(1))
print(round (43,6))
```

```python
# 모듈명이 길어서 축약으로 사용하고자 할 경우
import math as m
print(m.ceil(43.2)) # 올림정수 44
print(m.floor(43.2)) # 내림정수 43
print(m.sin(1))
```

```python
import random
print("# random 모듈")

#random():0.0 <= x < 1.0 사이의 float을 리턴.
print("- random():", random.random())

#uniform(min, max): 지정한 범위 사이의 float을 리턴
print("- uniform(10,20):", random.uniform(10, 20))

#randrange():지정한 범위의 int를 리턴.
# -randrange(max): 0부터 max사이의 값을 리턴
# -randrange(min,max): min부터 max사이의 값을 리턴
print("- randrange(10):",random.randrange(10))

#choice(list):리스트 내부에 있는 요소를 랜덤하게 선택
print("- choice[1,2,3,4,5]):",random.choice([1,2,3,4,5]))

#shuffle(list):리스트의 요소들을 랜덤하게 섞음.
print("- shuffle([1,2,3,4,5])",random.shuffle([1,2,3,4,5]))

#sample(list, k=<숫자>):리스트의 요소 중에 k개를 뽑음.
print("- sample([1,2,3,4,5])",random.sample([1,2,3,4,5], k=2))
```
